# Verify the representativeness of the sample dataset. #

```python
data[["JobRoleInterest", "CountryLive", "MoneyForLearning"]].head(10)
```

```python
job_interest_count = {
    "single_interest": 0,
    "multi_interest": 0
}

def job_interest(row):
    if "," in row:  
        job_interest_count["multi_interest"] += 1
    else:
        job_interest_count["single_interest"] += 1
        
JobRoleInterest = data["JobRoleInterest"].copy().dropna()
JobRoleInterest.apply(job_interest)

print(job_interest_count)
```

```python
job_interest_count_table = pd.Series(data=job_interest_count)

job_interest_count_table.plot.pie(autopct="%.1f%%", figsize=(6,6))
plt.ylabel("Single or Multi Interests")
plt.show()
```

```python
course_interest_count = {
    "web_or_mobile": 0,
    "others": 0
}

def course_interest(row):
    if "web" in row.lower() or "mobile" in row.lower():  
        course_interest_count["web_or_mobile"] += 1
    else:
        course_interest_count["others"] += 1

JobRoleInterest.apply(course_interest)

print(course_interest_count)
```

```python
course_interest_count_table = pd.Series(data=course_interest_count)

course_interest_count_table.plot.pie(autopct="%.1f%%", figsize=(6,6), startangle=-90)
plt.ylabel("Interested in Web or Mobile")
plt.show()
```

__Defining the data population:__ The purpose of our analysis is to find the 2 best markets for product advertisement. The company primarily provides web and mobile development although they do provide some other digital and tech courses as well. Assuming that we are aiming at new coders who only have learning experience less than 5 years, the data population should then include every coders with less than 5 years of learning experience who also are primarily interested in at least web or mobile developement around the world.

__Findings:__ According to the pie charts shown above, there is __6035 (86.3%) __ of people out of __6992__ who are interested in at least web or mobile. Additionally, __68.3%__ of people has multi-interest, while __31.7%__ has single interest. Based on the charts verification, the data can be a good representative to the population we are looking for since most new coders are interested in at least web or mobile development for both who have single and multi-interest.

# Analyse the amount participants willing to spend monthly. #

```python
data_InterestedRole[["MoneyForLearning", "MonthsProgramming"]].head(15)
```

```python
data_InterestedRole.loc[data_InterestedRole.MonthsProgramming == 0.0, ["MoneyForLearning", "MonthsProgramming"]].head(10)
```

__There are 0.0 month in `MonthsProgramming` column, which indicates that the participant has just started learning.__ To avoid dividing the `MoneyForLearning` by 0, replace all 0.0 with 1.0.

```python
data_InterestedRole["MonthsProgramming"].replace(
    to_replace=0.0,
    value=1.0,
    inplace=True
)
```

__Make sure the value is replaced.__

```python
data_InterestedRole.loc[data_InterestedRole.MonthsProgramming == 0.0, ["MoneyForLearning", "MonthsProgramming"]].head(10)
```

__Create new column for the money participants spent for learning monthly.__

```python
data_InterestedRole["MoneyForLearningMonthly"] = round(data_InterestedRole["MoneyForLearning"] / data_InterestedRole["MonthsProgramming"], 2)
```

```python
data_InterestedRole["MoneyForLearningMonthly"].head(10)
```

__Remove all null values.__

```python
remove_null = (pd.notnull(data_InterestedRole["MoneyForLearningMonthly"]))
```

```python
data_InterestedRole_MonthlySpend = data_InterestedRole[remove_null].copy()
data_InterestedRole_MonthlySpend["MoneyForLearningMonthly"].head(10)
```

```python
data_InterestedRole_MonthlySpend["CountryLive"].value_counts().head()
```

__Calculate the mean of monthly spending for the top 4 countries.__

```python
countries_focus = (data_InterestedRole_MonthlySpend.CountryLive == "United States of America") | (data_InterestedRole_MonthlySpend.CountryLive == "India") | (data_InterestedRole_MonthlySpend.CountryLive == "United Kingdom") | (data_InterestedRole_MonthlySpend.CountryLive == "Canada")
```

```python
data_InterestedRole_MonthlySpend[countries_focus].groupby(by="CountryLive")["MoneyForLearningMonthly"].describe()
```

__Calculate the median.__

```python
median = data_InterestedRole_MonthlySpend[countries_focus].groupby(by="CountryLive")["MoneyForLearningMonthly"].median()
print(median)
```

__Calculate the mode.__

```python
intervals = pd.interval_range(start=0, end=80000, freq=50)
intervals
```

```python
bins_MoneyForLearningMonthly = pd.Series(data=[0], index=intervals)
bins_MoneyForLearningMonthly.head()
```

```python
def mode_MoneyForLearningMonthly(col):
    sr = data_InterestedRole_MonthlySpend.loc[data_InterestedRole_MonthlySpend.CountryLive == col, "MoneyForLearningMonthly"]
    for r in sr:
        for interval in intervals:
            if r in interval:
                bins_MoneyForLearningMonthly.loc[interval] += 1
                break
    return bins_MoneyForLearningMonthly.sort_values(ascending=False)

mode_USA = mode_MoneyForLearningMonthly("United States of America")
mode_India = mode_MoneyForLearningMonthly("India")
mode_UK = mode_MoneyForLearningMonthly("United Kingdom")
mode_Canada = mode_MoneyForLearningMonthly("Canada")

mode_USA.plot.kde(label="USA", color="yellowgreen", linewidth=3, figsize=(9,6))
mode_India.plot.kde(label="India", color="gold", linewidth=3, figsize=(9,6))
mode_UK.plot.kde(label="UK", color="lightskyblue", linewidth=3, figsize=(9,6))
mode_Canada.plot.kde(label="Canada", color="lightcoral", linewidth=3, figsize=(9,6))

plt.legend()
```

__Findings:__ The table above shows that the mean is heavily affected by the outliers by comparing the mean, standard deviation and maximum value. If I do a quick estimation of z-scores for every row, it will definitely exceed 3 (any z-score beyond 3 means there is only 0.3% of proportion who is willing to pay more than 3 times of the standard deviation value adding from the mean). The shape of the kernel density estimation graph is right skewed and the main densities centred on range 0 to 500.

__Solution:__ Filter the range to amounts with z-score below 3. Choosing to remove outliers is appropriate in this context because the company's monthly subscriptions is only 59 dollar and we want to focus on the __common amounts most participants__ willing to pay. Alternatively, we can also count the number of participants who is willing to pay 59 dollar or more monthly.

__Validate outliers above 6000 for participants living in USA.__

```python
ls_cols_OutliersCheck = ["AttendedBootcamp", "BootcampFinish", "CountryLive", "MoneyForLearning", "MoneyForLearningMonthly", "MonthsProgramming", "HoursLearning", "SchoolDegree", "SchoolMajor"]
```

```python
data_4Countries.loc[data_4Countries.MoneyForLearningMonthly > 6000, ls_cols_OutliersCheck]
```

__Findings:__ To recap, the actual question of the survey was "Aside from university tuition, about how much money have you spent on learning to code so far (in US dollars)?".

Here is the list of possible items the participants spent on:

1. Event
2. Workshop 
3. Short course 
4. Online course 
5. Podcast 
6. School fee
7. bootcamp

The first table shows that there are only 5 participants who attended the bootcamp. Bootcamp is one of the costly investment for learning programming. However, there are 4 out of 5 participatns who had yet finish their bootcamp, which means they most likely made an upfront payment for the bootcamp, and thus the calculation for the `MoneyForLearningMonthly` column will not make sense since we are not dividing the spending by the number of months the upfront payment lasts until. Hence, my decision is to exclude any participant who only made an upfront payment but did not finish the bootcamp.

There is uncertainty regarding the 2 participants who finished their bootcamp but only learned programming for a month because bootcamp usually lasts for at least 2 months and a half. Since the information is uncertain, we will not include the 2 participants as well.

As for the rest who did not attend any bootcamp, options 1 to 5 are very unlikely because the costs are low. As for option 6, we know that the survey question excludes university tuition fee, thus option 6 is not valid as well. Since we are uncertain of what the participants spent that large amount of moeny on, hence, we will exclude participants who did not attend any bootcamp.

__Conclusion:__ We will remove those who:

1. Did not attend any bootcamp.
2. Did not complete the bootcamp.
3. Provide uncertain information (finished bootcamp but only learned programming for a month).
